```python
%config Completer.use_jedi = False
import matplotlib
import pickle
import glob
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sn
import sklearn.preprocessing as skp
import sklearn.metrics as skm
import sklearn.model_selection as skms
import tensorflow as tf
from matplotlib import gridspec

def rcparams(r=0.5):
    matplotlib.rcParams['font.size'] = 25 * r
    matplotlib.rcParams['xtick.labelsize'] = 20 * r
    matplotlib.rcParams['ytick.labelsize'] = 20 * r
    matplotlib.rcParams['axes.labelsize'] = 25 * r
    matplotlib.rcParams['legend.title_fontsize'] = 17 * r
    matplotlib.rcParams['legend.fontsize'] = 17 * r
    matplotlib.rcParams['axes.axisbelow'] = True
    matplotlib.rcParams['figure.figsize'] = [6 * r, 6 * r]

rcparams()

# format axis
def format_axis(ax):
    from matplotlib.ticker import (AutoMinorLocator)
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(which='both', width=2)
    ax.tick_params(which='major', length=6)
    ax.tick_params(which='minor', length=3)
    ax.locator_params(axis='x', nbins=5)
    ax.locator_params(axis='y', nbins=5)
    for axis in ['top', 'bottom', 'left', 'right']:
        ax.spines[axis].set_linewidth(1.5)
```

```python
PLOT_DIR = './'
```

### 1D CNN results

```python
%%time 
def result_1d():
    files = sorted(glob.glob(f'../results/pickle/mix-1-*-*-None.pickle'))
    acc = []
    for i, file in enumerate(files):
        with open(file, 'rb') as f:
            y_true = pickle.load(f)
            y_pred = pickle.load(f)
        acc.append(skm.accuracy_score(y_true, y_pred))
    return np.array(acc)


acc_1d = result_1d()
print(acc_1d)
```

### 2D CNN results

```python
%%time
def result_2d():
    accs = []
    for img_size in [50, 100, 150, 200, 250]:
        files = sorted(glob.glob(f'../results/pickle/mix-2-*-*-{img_size}.pickle'))
        acc = []
        for i, file in enumerate(files):
            with open(file, 'rb') as f:
                y_true = pickle.load(f)
                y_pred = pickle.load(f)
            acc.append(skm.accuracy_score(y_true, y_pred))
        accs.append(np.array(acc))
    return np.array(accs).squeeze()


acc_2d = result_2d()
print(acc_2d)
# each row is an image size
```

### Other machine learning results

```python
%%time
acc_1d = result_1d()[np.newaxis, ...]
acc_2d = result_2d()
acc_cnn = np.concatenate((acc_1d, acc_2d), axis=0).mean(axis=1)

# RBF-SVM 0.86143 RF 0.72571 k-NN: 0.65 GPC: 0.63286
acc = np.concatenate((acc_cnn, [0.86143, 0.72571, 0.65, 0.63286]))
print(acc)
```

```python
%%time
label = ['1D',
         '2D\n' + r'(50)',
         '2D\n' + r'(100)',
         '2D\n' + r'(150)',
         '2D\n' + r'(200)',
         '2D\n' + r'(250)',
         'SVM',
         'RF',
         'KNN',
         'GPC']

x = np.arange(0, len(acc), 1)

rcparams(0.75)
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(x, acc, capsize=4, width=0.6, color=['tab:gray'] * 4 + ['tab:red'] + ['tab:gray'] * 4)
ax.set_ylim([0.5, 1.0])
ax.grid(alpha=0.4)
format_axis(ax)
plt.xticks(x, label)
ax.set_ylabel('Accuracy')

# plt.savefig(PLOT_DIR + 'result-mix.svg', bbox_inches='tight')
plt.show()
```

### Confusion matrix 2D CNN size 200

```python
%%time
def load_2d_conf():
    conf, acc = [], []
    for k in range(1, 6):
        file = glob.glob(f'../results/pickle/mix-2-{k}-*-200.pickle')[0]
        with open(file, 'rb') as f:
            y_true = pickle.load(f)
            y_pred = pickle.load(f)
        conf.append(skm.confusion_matrix(y_true, y_pred))
        acc.append(skm.accuracy_score(y_true, y_pred))
    conf = np.array(conf).sum(axis=0) / 70
    
    conf = conf.T
    conf = np.flipud(conf)
    conf = pd.DataFrame(conf)
    return conf


conf = load_2d_conf()
```

```python
%%time
name = ['ABS', 'AC', 'PE', 'PET', 'PB', 'PC', 'PI',
        'PS', 'PP', 'PVC']

rcparams(0.65)
fig, ax = plt.subplots(figsize=(10, 8))
sn.heatmap(conf, annot=True, cmap="gray_r",
           square=True, vmax=1, vmin=0, fmt='0.2f', linewidths=0.25, linecolor='tab:gray')
ax.set_xlabel('True Species')
ax.set_ylabel('Predicted Species')
format_axis(ax)
ax.tick_params(which='minor', length=0)
plt.xticks(np.arange(0.5, 10, 1), name, horizontalalignment='center')
plt.yticks(np.arange(9.5, -0.5, -1), name, verticalalignment='center')

# plt.savefig(PLOT_DIR + 'result-confusion-matrix.scg', bbox_inches='tight')
plt.show()
```

### Load Example 1D data

```python
%%time
def load_1D_data():
    with open('../data/plastic.pickle', 'rb') as handle:
        X_train = pickle.load(handle)

    X_train = (X_train - X_train.min()) / (X_train.max() - X_train.min())

    y_train = np.concatenate((np.ones(70) * 0, np.ones(70),
                              np.ones(70) * 2, np.ones(70) * 3,
                              np.ones(70) * 4, np.ones(70) * 5,
                              np.ones(70) * 6, np.ones(70) * 7,
                              np.ones(70) * 8, np.ones(70) * 9,))

    X_train = np.random.RandomState(0).permutation(X_train)
    y_train = np.random.RandomState(0).permutation(y_train)

    X_train = X_train[..., np.newaxis]

    skf = skms.StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
    i = 0
    for train_index, test_index in skf.split(X_train, y_train):
        X_train_fold, X_test = X_train[train_index], X_train[test_index]
        y_train_fold, y_test = y_train[train_index], y_train[test_index]
    return X_test


X_1D = load_1D_data()
name = ['ABS', 'Acrylic', 'PE', 'PET', 'PB', 'PC', 'PI',
        'PS', 'PP', 'PVC']
```

### Saliency map function

```python
%%time
baseline = tf.zeros(shape=(200, 200, 2))


def saliency(plastic_id=0):
    gradSum = np.zeros((200, 200, 2))
    dataSum = np.zeros((200, 200, 2))
    oneDSum = np.zeros((4150,))
    for i in np.where(y_test == plastic_id)[0]:
        image = X_test[i].astype(np.float32)
        label = int(y_test[i])
        grad = ig(model, baseline, image, label, m_steps=50, batch_size=1)
        grad_abs = np.abs(grad.numpy().squeeze())
        grad_abs -= grad_abs.min()
        grad_abs /= grad_abs.max()
        gradSum += grad_abs
        dataSum += image
        oneDSum += X_1D[i, ..., 0]
    oneDSum = (oneDSum - oneDSum.min()) / (oneDSum.max() - oneDSum.min())
    return gradSum.mean(axis=2), dataSum, oneDSum
```

### Plot saliency funciton

```python
def plot_saliency(gradSum, dataSum, oneDSum, plastic_type=0, window=[]):
    fig = plt.figure(figsize=(8, 8))
    spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig, wspace=0.45, hspace=0.45)
    ax1 = fig.add_subplot(spec[0, 0])
    ax1.imshow(dataSum[..., 0], cmap="gray_r", origin="lower")
    plt.xticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    plt.yticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    ax1.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    ax1.set_ylabel(r"Wavenumber (cm$^{-1}$)")
    format_axis(ax1)
    plt.text(-60, 200, "(a)", va="center", ha="center", fontsize=22)

    ax2 = fig.add_subplot(spec[0, 1])
    ax2.imshow(dataSum[..., 1], cmap="gray_r", origin="lower")
    plt.xticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    plt.yticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    ax2.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    ax2.set_ylabel(r"Wavenumber (cm$^{-1}$)")
    format_axis(ax2)
    plt.text(-60, 200, "(b)", va="center", ha="center", fontsize=22)

    ax3 = fig.add_subplot(spec[1, 0])
    ax3.imshow(gradSum, cmap="gray_r", origin="lower")
    plt.xticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    plt.yticks([0, 50, 100, 150, 199], [2000, 2500, 3000, 3500, 4000])
    ax3.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    ax3.set_ylabel(r"Wavenumber (cm$^{-1}$)")
    format_axis(ax3)
    plt.text(-60, 200, "(c)", va="center", ha="center", fontsize=22)

    ax4 = fig.add_subplot(spec[1, 1])
    ax4.plot(np.linspace(2000, 4000, len(oneDSum)), oneDSum, 'k')
    for w in window:
        ax4.axvspan(w[0], w[1], alpha=0.5, color='tab:grey')
    ax4.set_xlim([2000, 4000])
    ax4.set_ylim([0, 1])
    ax4.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    ax4.set_ylabel(r"Normalized Intensity")
    format_axis(ax4)
    plt.text(-0.22, 1, "(d)", va="center", ha="center", fontsize=22, transform=ax4.transAxes)
    # plt.savefig(PLOT_DIR + f'sal_{name[plastic_type]}.svg', bbox_inches='tight')
    plt.show()
```
